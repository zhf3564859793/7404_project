```python
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.EState import Fingerprinter
from rdkit.Chem import MACCSkeys
fingerprint_list=[]
for j in drug['smile']:
    j1= Chem.MolFromSmiles(j)
    MACCS_fingerprint = MACCSkeys.GenMACCSKeys(j1)
    fingerprint_list.append(MACCS_fingerprint.ToBitString())
drug['MACCS_fingerprint']=fingerprint_list
drug
```

```python
#str to int
def trans_str_to_int(fingerprint_str):
    int_list=[]
    for ii in fingerprint_str:
        ii=int(ii)
        int_list.append(ii)
    return int_list
MACCS_fingerprint_list=[]
for s in drug['MACCS_fingerprint']:
    s1=trans_str_to_int(s)
    MACCS_fingerprint_list.append(s1)
drug['MACCS_fingerprint_int']=MACCS_fingerprint_list
drug
```

```python
Real_type_id_list=[]
for i in data_800_1000_compare['Real_DDI_Type']:
    Real_type_id = [j for j,x in enumerate(i) if x==1]
    Real_type_id_list.append(Real_type_id)
data_800_1000_compare['Real_type_id']=Real_type_id_list
data_800_1000_compare
```

```python
Predict_type_id_list=[]
for i in range(0,3438):
    pre=np.asarray(data_800_1000_compare['Predict_DDI_Type'][i])
    num=len(data_800_1000_compare['Real_type_id'][i])
    Predict_type_id=pre.argsort()[-1:-num-1:-1]
    Predict_type_id_list.append(Predict_type_id)
data_800_1000_compare['Predict_type_id']=Predict_type_id_list
data_800_1000_compare
```

```python
accuracy_rate_list=[]
for i in range(0,3438):
    real_set=set(data_800_1000_compare['Real_type_id'][i])
    pred_set=set(data_800_1000_compare['Predict_type_id'][i])
    same_number=len(real_set&pred_set)
    accuracy_rate=same_number/len(data_800_1000_compare['Real_type_id'][i])
    accuracy_rate_list.append(accuracy_rate)
data_800_1000_compare['accuracy_rate']=accuracy_rate_list
data_800_1000_compare
```

```python
data_800_1000_compare_sort=data_800_1000_compare.sort_values(by='accuracy_rate',ascending=False)
data_800_1000_compare_sort
```
