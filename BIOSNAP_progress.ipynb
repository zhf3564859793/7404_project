```python
drug_set=set(list4)
drug_list=list(drug_set)
drug=pd.DataFrame(drug_list)
drug
```

```python
import re
def strmove(STITCH):
    str_del_CID=re.sub('[CID]', '', STITCH)
    str_del_CID=str_del_CID.lstrip('0')
    return str_del_CID
drug_cid=[]
for i in drug[0]:
    cid=strmove(i)
    drug_cid.append(cid)
drug['cid']=drug_cid
drug
```

```python
import pubchempy as pcp
drug_smile=[]
for m in drug['cid']:
    print(m)
    c = pcp.Compound.from_cid(m)
    c_smile=c.isomeric_smiles
    drug_smile.append(c_smile)
    print(c_smile)
drug['smile']=drug_smile
drug
```
